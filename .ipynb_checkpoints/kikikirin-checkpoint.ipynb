```python
import os
from typing import List
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
import seaborn as sns
```

```python
font_files = font_manager.findSystemFonts(os.path.join(str(Path.home()), ".fonts"))

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

sns.set(font = "Noto Sans CJK JP")
```

```python
name = "きききりん"

df = pd.DataFrame(list(name), columns = ["letter"])
df["letter_count"] = 1
df = df.groupby("letter").sum()
df["ratio"] = df["letter_count"] / df["letter_count"].sum()
df
```

```python
ax = sns.barplot(data = df.reset_index(), x = "letter_count", y = "letter")
ax.set_title(f"{name}に含まれる文字数")
plt.show()
```
